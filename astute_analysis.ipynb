```python
def std(series):
    series = series.pct_change()
    return np.std(series)
```

```python
df.columns
```

```python
df_return = df.resample("Q-MAR").first().pct_change() * 100
df_std = df.resample("Q-MAR").agg(std) * 100
# df_return
fund = "ASTUTE QUANTUM FUND"
df_first_return = df_return[[fund]]
df_first_std = df_std[[fund]]
df_first_std.loc[df_first_std.idxmax(),:] = np.nan
print(df_first_std)
df_combine = df_first_return.join(df_first_std, lsuffix=" return", rsuffix=' std', how='inner').dropna(axis=0)
s = [size ** 1.8 for size in range(df_combine.shape[0])]

df_combine.plot(kind='scatter', x=f'{fund} std', y=f"{fund} return", s=s, figsize=(16,8), title=f"{fund} Snail Plot")
# df_std[df.columns[0]].iloc[1:]
# df_return[df.columns[0]].iloc[1:]
```
