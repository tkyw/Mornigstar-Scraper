```python
import pandas as pd
import numpy as np
```

```python
path = "/Users/tkyw/Documents/code/work"
filename = "/Malaysia Fund Universe (Formatted).xlsx"
file_path = path  + filename
```

```python
df = pd.read_excel(file_path, index_col = 0)
df
```

```python
df_new = pd.DataFrame(columns=["1Y Return", "3Y Return", "5Y Return", "10Y Return", "1Y Sharpe Ratio", "3Y Sharpe Ratio", "5Y Sharpe Ratio", "10Y Sharpe Ratio"])
```

```python
average_return = (1 + df.loc[:, ["Annualized Return" in col for col in df.columns]]) ** (1/12) - 1
average_return_1Y = average_return.iloc[:, -1]
annualized_return_1Y = (average_return_1Y + 1) ** (12) - 1
annualized_return_3Y = (average_return.iloc[:, -3:].mean(axis=1, skipna=False) + 1) ** (12 * 3) - 1
annualized_return_5Y = (average_return.iloc[:, -5:].mean(axis=1, skipna=False) + 1) ** (12 * 5) - 1
annualized_return_10Y = (average_return.iloc[:, -10:].mean(axis=1, skipna=False) + 1) ** (12 * 10) - 1

df_new['1Y Return'] = annualized_return_1Y
df_new['3Y Return'] = annualized_return_3Y
df_new['5Y Return'] = annualized_return_5Y
df_new['10Y Return'] = annualized_return_10Y

df_new
```

```python
sharpe_ratio = df.loc[:, ["Sharpe" in col for col in df.columns]]

sharpe_ratio_1Y = sharpe_ratio.iloc[:, -1:]
sharpe_ratio_3Y = sharpe_ratio.iloc[:, -3:].mean(axis=1, skipna=False)
sharpe_ratio_5Y = sharpe_ratio.iloc[:, -5:].mean(axis=1, skipna=False)
sharpe_ratio_10Y = sharpe_ratio.iloc[:, -10:].mean(axis=1, skipna=False)

df_new['1Y Sharpe Ratio'] = sharpe_ratio_1Y
df_new['3Y Sharpe Ratio'] = sharpe_ratio_3Y
df_new['5Y Sharpe Ratio'] = sharpe_ratio_5Y
df_new['10Y Sharpe Ratio'] = sharpe_ratio_10Y

df_new
```

```python
df_raw = pd.read_csv(path + "/Malaysia Fund Universe.csv", index_col=0)
ytd_sharpe = df_raw.loc["Sharpe 2024"].astype(float) / np.sqrt(12)
ytd_sharpe
```

```python
ytd_return = df_raw.loc["Average Return 2024", :].astype(float) / np.sqrt(12) * np.sqrt(3)
ytd_return
```

```python
df = pd.concat([df, df_new, ytd_sharpe, ytd_return], axis=1)
df
```

```python
df.rename(columns={"Average Return 2024": "YTD Return", "Sharpe 2024": "YTD Sharpe"}, inplace=True)
```

```python
columns = [col for col in df.columns if "Annualized" in col] + \
               ["YTD Return"] + \
                [col for col in df.columns if "Y R" in col] + \
                [col for col in df.columns if "Sharpe 20" in col or "YTD Sharpe" in col] + \
                [col for col in df.columns if "Sharpe Ratio" in col]
```

```python
df_combine = df.loc[:, columns]
df_combine.columns
```

```python
df_combine.to_excel("Malaysia Fund Universe (Updated).xlsx")
```
