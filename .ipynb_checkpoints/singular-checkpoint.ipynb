```python
def cumulative(df):
    df_rets = df.pct_change() + 1
    return (df_rets).prod() - 1

def calendar_year_return(df, lastest_date, last_date, first_year, w1=1, w2=0):
    df_new = pd.DataFrame(columns=['YTD', "1 Year", "3 Year", "5 Year", "7 Year", "10 Year", "Since Inception"] + list(map(str, list(reversed(range(2007, 2025))))))
    df_new["YTD"] = (df.loc[lastest_date.split("-")[0]].pct_change().dropna(axis=0)).sum()
    df_new["1 Year"] = (df.loc[pd.to_datetime(last_date):lastest_date].pct_change().dropna(axis=0)).sum()
    df_new["3 Year"] = (df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=3):lastest_date].pct_change().dropna(axis=0)).sum()
    df_new["5 Year"] = (df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=5):lastest_date].pct_change().dropna(axis=0)).sum()
    df_new["7 Year"] = (df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=7):lastest_date].pct_change().dropna(axis=0)).sum()
    df_new["10 Year"] = (df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=10):lastest_date].pct_change().dropna(axis=0)).sum()
    df_new["Since Inception"] = (df.loc[first_year:].pct_change()).sum()
    
    for year in list(reversed(range(2007, 2025))):
        df_new[str(year)] = (df.loc[str(year)].pct_change()).sum()
    
    # =====
    df_new.loc['Benchmark1', :] = df_new.loc['Benchmark1', :] * w1
    df_new.loc['Benchmark2', :] = df_new.loc['Benchmark2', :] * w2
    df_new.loc['Benchmark', :] =df_new.loc['Benchmark1', :] + df_new.loc['Benchmark2', :]
    # df_new['Benchmark2'] = df_new['Benchmark2'] * w2
    return df_new.iloc[[0, -1], :]

def semi_deviation(df, w1=1, w2=0):
    if w1 != 1:
        df.loc['Benchmark1', :] = df.loc['Benchmark1', :] * w1
        df.loc['Benchmark2', :] = df.loc['Benchmark2', :] * w2
        df.loc['Benchmark', :] = df.loc['Benchmark1', :] + df.loc['Benchmark2', :]
        df = df.iloc[[0, -1], :]
    msq = (df[df < 0] ** 2).sum()
    # print(df)
    sd = (msq / (df.shape[0] - 1)) ** 0.5
    return sd
    
        
    
def downside_risk(df, lastest_date, last_date, first_year, w1=1, w2=0):
    df_new = pd.DataFrame(columns=['YTD', "1 Year", "3 Year", "5 Year", "7 Year", "10 Year", "Since Inception"] + list(map(str, list(reversed(range(2007, 2025))))))
    df_new["YTD"] = semi_deviation((df.loc[lastest_date.split("-")[0]].dropna(axis=0).pct_change().dropna(axis=0)), w1=w1,w2=w2) * np.sqrt(252)
    df_new["1 Year"] = semi_deviation((df.loc[pd.to_datetime(last_date):lastest_date].dropna(axis=0).pct_change().dropna(axis=0)), w1=w1,w2=w2) * np.sqrt(252)
    df_new["3 Year"] = semi_deviation((df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=3):lastest_date].dropna(axis=0).pct_change().dropna(axis=0)), w1=w1,w2=w2) * np.sqrt(252)
    df_new["5 Year"] = semi_deviation((df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=5):lastest_date].dropna(axis=0).pct_change().dropna(axis=0)), w1=w1,w2=w2) * np.sqrt(252)
    df_new["7 Year"] = semi_deviation((df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=7):lastest_date].dropna(axis=0).pct_change().dropna(axis=0)), w1=w1,w2=w2) * np.sqrt(252)
    df_new["10 Year"] = semi_deviation((df.loc[pd.to_datetime(last_date) - pd.DateOffset(years=10):lastest_date].dropna(axis=0).pct_change().dropna(axis=0)), w1=w1,w2=w2) * np.sqrt(252)
    df_new["Since Inception"] = semi_deviation((df.loc[first_year:].pct_change()), w1=w1,w2=w2) * np.sqrt(252)
    
    for year in list(reversed(range(2007, 2025))):
        try:
            df_new[str(year)] = semi_deviation(df.loc[str(year)].pct_change(), w1=w1,w2=w2) * np.sqrt(252)
        except:
            print(df_new[str(year)])
    
    return df_new
        
```

## Upside Capture

```python
def geomean(df):
    df += 1
    return (df.prod() ** (1 / df.shape[0])) - 1
```

```python
df_active = df_new.loc["2023-01-3": "2024-01-02"].dropna(axis=0).resample("M").first().pct_change().dropna(axis=0)
# df_active = df_new.loc["2024-01-02":].dropna(axis=0).resample("M").first().pct_change().dropna(axis=0)
upside = df_active.query("Price > 0")
upside_geo = geomean(upside)
upside_capture = upside_geo['Last'] / upside_geo['Price']
downside = df_active.query("Price <= 0")
downside_geo = geomean(downside)
downside_capture = downside_geo['Last'] / downside_geo['Price']
capture_ratio = upside_capture / downside_capture
copy(pd.Series([upside_capture, downside_capture, capture_ratio]), sort=False)
```

```python
years = list(reversed(range(2010, 2024)))
index = ["Upside", "Downside", "Capture"]
captures = pd.DataFrame(columns=years, index=index)
for year in years:
        df_active = df_new.loc[str(year)].dropna(axis=0).resample("M").first().pct_change().dropna(axis=0)
        upside = df_active.query("Price > 0")
        upside_geo = geomean(upside)
        upside_capture = upside_geo['Last'] / upside_geo['Price']
        downside = df_active.query("Price <= 0")
        downside_geo = geomean(downside)
        downside_capture = downside_geo['Last'] / downside_geo['Price']
        capture_ratio = upside_capture / downside_capture
        captures[year] = [upside_capture, downside_capture, capture_ratio]
copy(captures, sort=False, index=True)
```
