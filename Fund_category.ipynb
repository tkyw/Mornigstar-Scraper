```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Malaysia Fund Universe (Updated) Alpha.xlsx", sheet_name=['Sheet1'], index_col=0, na_values=['-inf', '-'])['Sheet1']
df
```

```python
def group_category(s):
    try:
        if "Equity" in s:
            return "Equity"
        elif "Balanced" in s:
            return "Balanced"
        elif "Bond" in s:
            return "Fixed Income"
        else:
            return "Others"
    except:
        return "Others"
```

```python
df['CategoryName']
```

```python
df['Category_broad'] = df['CategoryName'].apply(group_category)
df
```

```python
groupby_df = df.groupby(["Category_broad", "Name"])[['1Y Return', "3Y Return", "5Y Return", "10Y Return", "1Y Sharpe Ratio", "3Y Sharpe Ratio", "5Y Sharpe Ratio", "10Y Sharpe Ratio", "AlphaM36"]].agg(np.median)
idx = pd.IndexSlice

# groupby_df.loc[idx['Balanced', :], "1Y Return"]
list(set(groupby_df.index.get_level_values(0)))
groupby_df.loc[idx["Equity", :], :].sort_values(['10Y Return', "5Y Return", "3Y Return", "1Y Return"], ascending=False).to_clipboard(excel=True)
```
